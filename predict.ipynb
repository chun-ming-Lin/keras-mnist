```python
# 用小畫家3D 繪製圖片測試
import cv2

for i in range(10):
    img = cv2.imread(str(i)+ '.jpg', 0)
    pred = model.predict_classes((255 - img.reshape(1, 28*28))/255)
    print(pred)

```
